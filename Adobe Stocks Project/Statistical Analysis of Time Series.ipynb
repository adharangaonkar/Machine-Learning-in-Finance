```python
from scipy import stats
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas_datareader import DataReader
from datetime import datetime
plt.style.use('fivethirtyeight')
import operator
```

```python
ticker = 'ADBE'
start = datetime(2019,1,1)
end = datetime(2021,4,1)
df = DataReader(ticker,  'yahoo', start, end)
```

```python
df
```

```python
a = df.Close[1:].tolist()
```

```python
b = df.Close[:-1].tolist()
```

```python
c = np.subtract(a,b)
```

```python
ret = np.true_divide(c, b)
```

```python
ret
```

```python
print(round(np.mean(ret),5))
```

```python
print(round(np.mean(ret),5), stats.ttest_1samp(ret,0))
```

The T-value is 1.65 while the P-value is 0.09. Thus, the mean is
statistically not different from zero.

trading strategy by taking a long position if today's price is close to the minimum

is close to its 52-week high. The following Python program presents this 52-week's
range and today's position:
